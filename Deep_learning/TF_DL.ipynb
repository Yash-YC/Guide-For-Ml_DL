```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create features
X = np.array([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])

# Create labels
y = np.array([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])

# Visualize it
plt.scatter(X, y)
```

```python
X = tf.cast(tf.constant(X), dtype = tf.float32)
y = tf.cast(tf.constant(y), dtype = tf.float32)
```

```python
# steps in model with tensorflow 
# set a random seed
tf.random.set_seed(42)

# create a model 
model = tf.keras.Sequential([layers.Dense(1)])

#compile a model 
model.compile(tf.keras.optimizers.SGD(),
             keras.losses.mae,
             metrics= ["mae"])

#fit the model
model.fit(X,y,epochs = 5)
```

```python
model = tf.keras.Sequential()
model.add(layers.Dense(1))

#compile a model 
model.compile(tf.keras.optimizers.SGD(),
             keras.losses.mae,
             metrics= ["mae"])

#fit the model
model.fit(X,y,epochs = 5)
```

```python
X,y
```

```python
model.predict([17.0])
```

```python
# Imporving the model
model = tf.keras.Sequential()
model.add(layers.Dense(50, "relu"))
model.add(layers.Dense(1))

#compile a model 
model.compile(tf.keras.optimizers.Adam(lr = 0.01),
             keras.losses.mae,
             metrics= ["mae"])

#fit the model
model.fit(X,y,epochs = 100)
```

```python
model.predict([17.0])
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model = model ,show_shapes= True)
```

```python
import pandas as pd
data = pd.read_csv("insurance.csv") 
```

```python
data
```

```python
# one hot encode
insurance = pd.get_dummies(data)
insurance
```

```python
X = insurance.drop("charges", axis = 1)
y = insurance["charges"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size= 0.33 ,random_state= 42 )
len(X)
```

```python
tf.random.set_seed(42)
i_model = tf.keras.Sequential()
i_model.add(layers.Dense(100))
i_model.add(layers.Dense(10))
i_model.add(layers.Dense(1))

i_model.compile(loss = "mae", optimizer =tf.keras.optimizers.Adam(), metrics= "mae")

history = i_model.fit(X_train,y_train ,epochs = 300)
```

```python
#check result
i_model.evaluate(X_test,y_test)
```

```python
i_model.summary()
```

```python
pd.DataFrame(history.history).plot()
```

```python
# Normalization and Scandardization
# in terms of Scaling values , neural networks tend to prefer normalization 
# But of you are not sure you should try both and see which Perform Better 
```

```python
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
from sklearn.compose import make_column_transformer
Scalar = MinMaxScaler()
OHE = OneHotEncoder()

ct = make_column_transformer(
    (Scalar ,["age","bmi", "children"]),
    (OHE ,["sex","smoker","region"])
)

X = data.drop("charges", axis = 1)
y = data["charges"]


X_train,X_test,y_train,y_test = train_test_split(X,y,train_size= 0.2 ,random_state= 42 )

# fit
ct.fit(X_train)

X_train_normal = ct.transform(X_train)
X_test_normal = ct.transform(X_test)
```

```python
X_test_normal.shape
```

```python
tf.random.set_seed(42)

model1 = tf.keras.Sequential()
model1.add(layers.Dense(100))
model1.add(layers.Dense(10))
model1.add(layers.Dense(1))

model1.compile(loss = "mae",
                optimizer = tf.keras.optimizers.Adam(),
                metrics= "mae" )

model1.fit(X_train_normal,y_train ,epochs = 300)
```

```python
model1.evaluate(X_test_normal,y_test)
```
