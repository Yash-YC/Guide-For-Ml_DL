## **Confusion Matrix**

**True Positive (TP)** 
The predicted value matches the actual value
The actual value was positive and the model predicted a positive value

**True Negative (TN)** 
The predicted value matches the actual value
The actual value was negative and the model predicted a negative value

**False Positive (FP)  Type 1 error**
The predicted value was falsely predicted
The actual value was negative but the model predicted a positive value
Also known as the Type 1 error

**False Negative (FN)  Type 2 error**
The predicted value was falsely predicted
The actual value was positive but the model predicted a negative value
Also known as the Type 2 error

```python
actual_values = [1, 1, 0, 1, 0, 0, 1, 0, 0, 0]
predictions =   [1, 0, 1, 1, 1, 0, 1, 1, 0, 0]
```

## **Accuracy Pradox**

Accuracy is not a reliable metric to determine a model performance for imbalanced data.

Thats why its called a Paradox because, intuitively, youd expect a Model with a higher Accuracy to have been the best Model but Accuracy Paradox tells us that this, sometimes, isnt the case.

So we use F1 score: 

**2 ((precision*recall)/(precision+recall))** 

## **ROC Curve (Receiver operating characteristic)**

The ROC curve is the plot between sensitivity and **(1- specificity). (1- specificity)** is also known as false positive rate and sensitivity is also known as True Positive rate.


```python
Image(filename="Images/c3.png")
```

```python
print('AUC-ROC:', roc_auc_score(actual_values, predictions))
```

```python
print('AUC-ROC:', roc_curve(actual_values, predictions))
```
